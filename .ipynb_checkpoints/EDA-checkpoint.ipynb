we will need some visualization tools for our experiments

```python
from bokeh.io import output_file, show ,output_notebook 
from bokeh.layouts import column
from bokeh.plotting import figure
output_notebook()
from bokeh.layouts import gridplot
import matplotlib.pyplot as plt
import seaborn as sns
```

in this experiments we goin to check train test sizes effect on all of our algotirhems 
* we will choose random preference for our algorithems 
* the only change will be the train test size

in our data we will try train size from 70% to 90%  
in jumps of 1%

our commom preferance

```python
model = PickleHandler.load_model('models/train_test_size_1')
print_d(model['dataframe'])
print_d(model['algorithm'])
```

we are going to process all of these modeles if a for loop

now we going to get our results

```python
model_lst = []
res_lst = []
for i in range(21): model_lst.append(PickleHandler.load_model(f'models/train_test_size_{i+1}'))
for i in model_lst:res_lst.append(i['res'])
```

```python
algo_dict = {'my tree':[],'my NB':[],'sklearn NB':[],'knn':[],'sklearn tree':[],'K_means':[]}
for j in algo_dict:
    train,test,mean = [],[],[]
    for i in range(21):
        a = float(res_lst[i][j]['train on train']['accuracy_score'])
        b = float(res_lst[i][j]['train on test']['accuracy_score'])
        train.append(a)
        test.append(b)
        mean.append((a+b)/2)
    algo_dict[j] = [mean , train , test]
```

```python
x = []
for i in range(21):
    x.append((70 + i)/100)
```

```python
p= list()
for i,count in zip(algo_dict,range(6)):   
    p.append(figure(y_axis_label=i))
    p[count].line(x, algo_dict[i][0] ,legend_label=" mean", line_width=4 ,  color="red")
    p[count].line(x, algo_dict[i][1] ,legend_label=" train on train", color="black",line_width=2)
    p[count].line(x, algo_dict[i][2] ,legend_label=" train on test", line_width=2,color="blue")
    p[count].legend.location = "bottom_right"
    p[count].legend.click_policy="hide"

    
grid = gridplot([p[0:1], p[1:2], p[2:3], p[3:4], p[4:5], p[5:6]], width=700, height=250)
show(grid)
```

We can see clearly that there is some bug in our decision tree at train size = 76%

```python
index_avg = 0 
print('mean score')
for i in algo_dict:
    max_val = max(algo_dict[i][0])
    max_index = algo_dict[i][0].index(max_val)
    index_avg += x[max_index]
    print(f'{i} max pick at {x[max_index]}' )
```

```python
print('train on train score')
for i in algo_dict:
    max_val = max(algo_dict[i][1])
    max_index = algo_dict[i][1].index(max_val)
    index_avg += x[max_index]
    print(f'{i} max pick at {x[max_index]}' )
```

```python
print('train on test score')
for i in algo_dict:
    max_val = max(algo_dict[i][2])
    max_index = algo_dict[i][2].index(max_val)
    index_avg += x[max_index]
    print(f'{i} max pick at {x[max_index]}' )
```

# Second experiment - K-Means clustes num

```python
from sklearn.cluster import KMeans
```

We saw that our k-mean classifier went a little crazy in our last experiment let's find for him the best cluster nums

but for real we need to look for the elbow

```python
model_lst = []
res_lst = []
for i in range(21): model_lst.append(PickleHandler.load_model(f'models/K-Means_test{i+1}'))
for i in model_lst:res_lst.append(i['res']['K_means'])
```

```python
train,test,mean = [],[],[]
for i in range(21):
        a = float(res_lst[i]['train on train']['accuracy_score'])
        b = float(res_lst[i]['train on test']['accuracy_score'])
        train.append(a)
        test.append(b)
        mean.append((a+b)/2)
```

```python
x = range(1,22)
p= figure(y_axis_label="modeles accuracy score",x_axis_label= "train size")


p.line(x,mean,legend_label='mean', line_width=4 , color = 'red' )
p.line(x,train,legend_label='train', line_width=4 , color = 'blue' )
p.line(x,test,legend_label='test', line_width=4 , color = 'green' )
p.legend.location = "bottom_right"
p.legend.click_policy="hide"
show(p)
```

In our algorithm process we get a x_train file to check on the same x_train the Variance

```python
x_train = pd.read_csv("models/x_train_k_means_test.csv")
```

```python
y =[]
for i in range(21):
    kmeans = KMeans(n_clusters=i+1, random_state=0).fit(x_train)
    temp = kmeans.inertia_
    y.append(temp)
```

```python
p= figure(y_axis_label="Variance",x_axis_label= "number of clusters")
p.line(x,y, line_width=4 )
show(p)
```
